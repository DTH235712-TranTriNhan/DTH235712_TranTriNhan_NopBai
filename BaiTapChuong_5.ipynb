```python
# Câu 4: Các hàm quan trọng trong xử lý chuỗi của Python
'''
1. Các hàm cơ bản
Hàm / Phương thức	Chức năng	Ví dụ
len(s)	Trả về độ dài chuỗi	len("Hello") → 5
str()	Chuyển giá trị khác (số, bool, v.v.) sang chuỗi	str(123) → "123"
type(s)	Kiểm tra kiểu dữ liệu	type("abc") → <class 'str'>

2. Loại bỏ / chỉnh sửa khoảng trắng
Phương thức	Mô tả	Ví dụ
s.strip()	Xóa khoảng trắng ở đầu và cuối chuỗi	" hi ".strip() → "hi"
s.lstrip()	Xóa khoảng trắng ở bên trái	" hi ".lstrip() → "hi "
s.rstrip()	Xóa khoảng trắng ở bên phải	" hi ".rstrip() → " hi"

3. Cắt và ghép chuỗi
Phương thức	Mô tả	Ví dụ
s.split(sep)	Cắt chuỗi thành danh sách theo ký tự phân tách	"a,b,c".split(',') → ['a','b','c']
' '.join(list)	Ghép danh sách thành chuỗi, cách nhau bởi ' '	' '.join(['Xin','chào']) → "Xin chào"

4. Chuyển đổi chữ hoa – thường
Phương thức	Mô tả	Ví dụ
s.upper()	Chuyển sang chữ hoa	"abc".upper() → "ABC"
s.lower()	Chuyển sang chữ thường	"ABC".lower() → "abc"
s.title()	Viết hoa chữ cái đầu của mỗi từ	"xin chao ban".title() → "Xin Chao Ban"
s.capitalize()	Viết hoa chữ cái đầu tiên của chuỗi	"hello world".capitalize() → "Hello world"

5. Tìm kiếm và thay thế
Phương thức	Mô tả	Ví dụ
s.find(sub)	Trả về vị trí đầu tiên tìm thấy sub (hoặc -1 nếu không có)	"hello".find("l") → 2
s.index(sub)	Giống find() nhưng lỗi nếu không tìm thấy	"hello".index("l") → 2
s.count(sub)	Đếm số lần xuất hiện của chuỗi con	"banana".count("a") → 3
s.replace(old, new)	Thay thế chuỗi old bằng new	"hi bob".replace("bob","alice") → "hi alice"

6. Kiểm tra nội dung chuỗi
Phương thức	Mô tả	Ví dụ
s.isalpha()	Chỉ gồm chữ cái?	"abc".isalpha() → True
s.isdigit()	Chỉ gồm số?	"123".isdigit() → True
s.isalnum()	Gồm chữ và số?	"abc123".isalnum() → True
s.isspace()	Chỉ gồm khoảng trắng?	" ".isspace() → True
s.startswith(prefix)	Bắt đầu bằng chuỗi con?	"hello".startswith("he") → True
s.endswith(suffix)	Kết thúc bằng chuỗi con?	"hello".endswith("lo") → True

'''
```

```python
# Câu 5: Xử lý chuỗi với các hàm cơ bản
# Nhập chuỗi từ người dùng
s = input("Nhập vào một chuỗi: ")

# Khởi tạo các biến đếm
hoa = thuong = so = dacbiet = khoangtrang = nguyenam = phuam = 0

# Tập hợp các nguyên âm (cả hoa và thường)
nguyen_am_set = "aeiouAEIOU"

# Duyệt từng ký tự trong chuỗi
for ch in s:
    if ch.isupper():
        hoa += 1
    elif ch.islower():
        thuong += 1
    elif ch.isdigit():
        so += 1
    elif ch.isspace():
        khoangtrang += 1
    else:
        dacbiet += 1

    # Kiểm tra nguyên âm / phụ âm
    if ch.isalpha():  # Nếu là chữ cái
        if ch in nguyen_am_set:
            nguyenam += 1
        else:
            phuam += 1

# Xuất kết quả
print("Chuỗi bạn nhập:", s)
print("Số chữ in HOA:", hoa)
print("Số chữ in thường:", thuong)
print("Số chữ là chữ số:", so)
print("Số ký tự đặc biệt:", dacbiet)
print("Số ký tự khoảng trắng:", khoangtrang)
print("Số chữ nguyên âm:", nguyenam)
print("Số chữ phụ âm:", phuam)
```

```python
# Câu 6: Trích lọc số âm trong chuỗi
import re

def NegativeNumberInStrings(s):
    # Tìm tất cả chuỗi có dạng: dấu '-' theo sau là 1 hoặc nhiều chữ số
    numbers = re.findall(r'-(\d+)', s)
    
    # Chuyển các chuỗi số tìm được thành số nguyên âm
    negatives = [-int(num) for num in numbers]
    
    # Xuất kết quả
    print("Các số nguyên âm trong chuỗi là:", *negatives)

# --- Ví dụ ---
NegativeNumberInStrings("abc-5xyz-12k9l--p")
```

```python
# Câu 7: Tối ưu chuỗi danh từ
def ToiUuChuoiDanhTu(s):
    # B1: Xóa khoảng trắng ở đầu và cuối
    s = s.strip()
    
    # B2: Cắt chuỗi thành các từ (tự loại bỏ khoảng trắng dư thừa)
    arr = s.split()
    
    # B3: Viết hoa chữ cái đầu mỗi từ, còn lại viết thường
    arr = [word.capitalize() for word in arr]
    
    # B4: Ghép các từ lại thành 1 chuỗi, cách nhau bởi 1 khoảng trắng
    result = ' '.join(arr)
    
    return result

# --- Ví dụ sử dụng ---
chuoi = input("Nhập chuỗi danh từ: ")
print("Chuỗi tối ưu:", ToiUuChuoiDanhTu(chuoi))
```

```python
#Câu 8: Tách lấy tên bài hát


import os

# Hàm 1: Lấy tên file có phần mở rộng
def LayTenFile(dayduongdan):
    return os.path.basename(dayduongdan)

# Hàm 2: Lấy tên file không có phần mở rộng
def LayTenFileKhongDuoi(dayduongdan):
    tenfile = os.path.basename(dayduongdan)
    ten_khong_duoi, _ = os.path.splitext(tenfile)
    return ten_khong_duoi

# --- Ví dụ ---
duongdan = "d:\\music\\nhandeptrai.mp3"
print("Tên file có đuôi:", LayTenFile(duongdan))
print("Tên file không có đuôi:", LayTenFileKhongDuoi(duongdan))
```
