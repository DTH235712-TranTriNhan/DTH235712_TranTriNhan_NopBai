```python
# Câu 6: Nhập một số n có tối đa 2 chữ số. Hãy cho biết cách đọc ra dạng chữ.
# (vd: n=35 => Ba mươi lăm, n=5 => năm). 
don_vi = {
    0: 'không', 1: 'một', 2: 'hai', 3: 'ba', 4: 'bốn',
    5: 'năm', 6: 'sáu', 7: 'bảy', 8: 'tám', 9: 'chín'
}

hang_chuc = {
    1: 'mười', 2: 'hai mươi', 3: 'ba mươi', 4: 'bốn mươi',
    5: 'năm mươi', 6: 'sáu mươi', 7: 'bảy mươi', 8: 'tám mươi',
    9: 'chín mươi'
}

def doc_so_hai_chu_so(n):
    """
    Đọc một số nguyên n có tối đa 2 chữ số ra dạng chữ.
    Giả định n là một số nguyên không âm.
    """
    if n < 0 or n > 99:
        return "Số không hợp lệ. Vui lòng nhập số từ 0 đến 99."

    # Trường hợp số từ 0 đến 9
    if n < 10:
        return don_vi[n]

    # Trường hợp số từ 10 đến 19
    if n < 20:
        if n == 10:
            return "mười"
        chu_so_don_vi = n % 10
        if chu_so_don_vi == 5:
            return "mười lăm"
        return f"mười {don_vi[chu_so_don_vi]}"

    # Trường hợp số từ 20 đến 99
    chu_so_chuc = n // 10
    chu_so_don_vi = n % 10

    phan_chuc = hang_chuc[chu_so_chuc]
    phan_don_vi = ""

    if chu_so_don_vi == 0:
        phan_don_vi = ""
    elif chu_so_don_vi == 1:
        phan_don_vi = "mốt"
    elif chu_so_don_vi == 4:
        phan_don_vi = "tư"
    elif chu_so_don_vi == 5:
        phan_don_vi = "lăm"
    else:
        phan_don_vi = don_vi[chu_so_don_vi]

    if phan_don_vi:
        return f"{phan_chuc} {phan_don_vi}"
    else:
        return phan_chuc

# Ví dụ sử dụng
so_nhap = int(input("Mời bạn nhập một số có tối đa 2 chữ số (0-99): "))
ket_qua = doc_so_hai_chu_so(so_nhap)
print(f"Số {so_nhap} đọc là: {ket_qua}")
```

```python
# Câu 7: Nhập vào một ngày (ngày, tháng, năm). Tìm ngày kế sau ngày
# vừa nhập (ngày/tháng/năm).
def is_valid_date(day, month, year):
    """
    Kiểm tra tính hợp lệ của ngày, tháng, năm nhập vào.
    """
    if not (1 <= month <= 12):
        return False

    days_in_month = [0, 31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]

    if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
        days_in_month[2] = 29

    if not (1 <= day <= days_in_month[month]):
        return False

    return True

def find_next_day(day, month, year):
    """
    Tìm ngày kế tiếp của một ngày cho trước.
    """
    days_in_month = [0, 31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]

    if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
        days_in_month[2] = 29

    day += 1

    if day > days_in_month[month]:
        day = 1
        month += 1
        if month > 12:
            month = 1
            year += 1

    return day, month, year

try:
    day = int(input("Nhập ngày: "))
    month = int(input("Nhập tháng: "))
    year = int(input("Nhập năm: "))

    if not is_valid_date(day, month, year):
        print("Ngày/tháng/năm bạn nhập không hợp lệ. Vui lòng thử lại.")
    else:
        next_day, next_month, next_year = find_next_day(day, month, year)
        print(f"Ngày kế tiếp là: {next_day}/{next_month}/{next_year}")

except ValueError:
    print("Dữ liệu nhập vào không hợp lệ. Vui lòng nhập số nguyên.")
```

```python
# Câu 18: Vẽ các hình dưới đây
def draw_shape1(n):
    print("Hình 1:")
    for i in range(n):
        for j in range(n):
            if i == 0 or i == n-1 or j == 0 or j == n-1:
                print("*", end=" ")
            else:
                print(" ", end=" ")
        print()
    print()


def draw_shape2(n):
    print("Hình 2:")
    for i in range(1, n + 1):
        print("* " * i)
    print()


def draw_shape3(n):
    print("Hình 3:")
    for i in range(n):
        for j in range(n):
            if j == 0 or i == 0 or i == j:
                print("*", end=" ")
            else:
                print(" ", end=" ")
        print()
    print()


n = int(input("Nhập chiều cao n: "))
draw_shape1(n)
draw_shape2(n)
draw_shape3(n)
```

```python
# Câu 19: Tính giá trị biểu thức S
import math

def tinh_giai_thua(k):
    """Hàm tính giai thừa của một số nguyên không âm k."""
    if k == 0 or k == 1:
        return 1
    giai_thua = 1
    for i in range(2, k + 1):
        giai_thua *= i
    return giai_thua

def tinh_bieu_thuc_S(x, n):
    """Hàm tính giá trị của biểu thức S(x,n)."""
    tong_S = 0
    
    # Vòng lặp tính tổng các số hạng
    for i in range(n + 1):
        tu_so = x**(2 * i + 1)
        mau_so = tinh_giai_thua(2 * i + 1)
        
        so_hang = tu_so / mau_so
        tong_S += so_hang
        
    return tong_S

# Nhập dữ liệu từ người dùng
try:
    x = float(input("Nhập giá trị của x: "))
    n = int(input("Nhập giá trị của n: "))

    if n < 0:
        print("n phải là số nguyên không âm.")
    else:
        ket_qua = tinh_bieu_thuc_S(x, n)
        print(f"Giá trị của biểu thức S là: {ket_qua}")

except ValueError:
    print("Dữ liệu nhập vào không hợp lệ. Vui lòng nhập số.")
```
