```python
# Câu 4: Viết Hàm tính ROI
def ROI(dt,cp):
 return (dt-cp)/cp
def GoiYDauTu(roi):
 if roi>=0.75:
  return "Nên đầu tư"
 else:
  return "Không nên đầu tư"
print("Chương trình tính ROI")
dt=int(input("Nhập Doanh Thu:"))
cp=int(input("Nhập chi phí:"))
roi=ROI(dt,cp)
print("Tỉ Lệ ROI=",roi)
print("==>",GoiYDauTu(roi))
```

```python
# Câu 7: Tính và xuất độ dài đoạn AB
import math

# Nhập tọa độ điểm A
xA = float(input("Nhập tọa độ x của điểm A: "))
yA = float(input("Nhập tọa độ y của điểm A: "))

# Nhập tọa độ điểm B
xB = float(input("Nhập tọa độ x của điểm B: "))
yB = float(input("Nhập tọa độ y của điểm B: "))

# Tính độ dài đoạn thẳng AB
do_dai_AB = math.sqrt((xB - xA)**2 + (yB - yA)**2)

# In kết quả
print(f"Độ dài đoạn thẳng AB là: {do_dai_AB}")
```

```python
# Câu 8: Viết chương trình tính loga
import math

# Vòng lặp để đảm bảo người dùng nhập đúng điều kiện
while True:
    try:
        a = float(input("Nhập cơ số a (a > 0 và a != 1): "))
        x = float(input("Nhập giá trị x (x > 0): "))

        # Kiểm tra điều kiện của a và x
        if a > 0 and a != 1 and x > 0:
            # Tính logarit bằng công thức đổi cơ số
            log_a_x = math.log(x) / math.log(a)
            print(f"Logarit cơ số {a} của {x} là: {log_a_x}")
            break  # Thoát vòng lặp khi nhập đúng
        else:
            print("Giá trị a và x không hợp lệ. Vui lòng nhập lại.")
    except ValueError:
        print("Đầu vào không hợp lệ. Vui lòng nhập một số.")
```

```python
# Câu 9: Viết chương trình tính căn bậc 2 lồng nhau
import math

def tinh_can_bac_2_long_nhau(x, so_lan_lap):
    """
    Hàm tính giá trị của căn bậc hai lồng nhau.

    Args:
        x (float): Giá trị bên trong căn (x >= 0).
        so_lan_lap (int): Số lần lồng nhau.
    
    Returns:
        float: Giá trị xấp xỉ của biểu thức.
    """
    
    if x < 0:
        return "Lỗi: x phải là số không âm."
    
    # Bắt đầu từ số 0 để tính từ trong ra
    ket_qua = 0
    
    # Lặp lại 'so_lan_lap' lần
    for _ in range(so_lan_lap):
        ket_qua = math.sqrt(x + ket_qua)
        
    return ket_qua

# --- Phần nhập liệu và chạy chương trình ---

try:
    x = float(input("Nhập giá trị x (x >= 0): "))
    so_lan_lap = int(input("Nhập số lần lồng nhau (số nguyên dương): "))
    
    # Gọi hàm và in kết quả
    ket_qua_cuoi = tinh_can_bac_2_long_nhau(x, so_lan_lap)
    print(f"Giá trị của biểu thức căn bậc hai lồng nhau là: {ket_qua_cuoi}")

except ValueError:
    print("Đầu vào không hợp lệ. Vui lòng nhập một số.")
```

```python
# Câu 12: Hàm oscillate
def oscillate(start, stop):
    """
    Tạo ra một chuỗi số dao động theo quy luật:
    - Giá trị ban đầu, sau đó giá trị tuyệt đối.
    - Các cặp số (i, -i) lặp lại.
    """
    if start <= 0:
        yield start
        # Tạo cặp số (i, -i) cho các giá trị âm
        for i in range(start + 1, 0):
            yield abs(i)  # Giá trị dương
            yield i       # Giá trị âm
    else: # Trường hợp start > 0
        yield start
        
    # Tạo cặp số (i, -i) cho các giá trị từ 0 đến stop
    for i in range(0, stop):
        yield i
        yield -i

# --- Kiểm tra hàm ---
for n in oscillate(-3, 5):
    print(n, end=' ')
print()

# Kết quả mong muốn: -3 3 -2 2 -1 1 0 0 1 -1 2 -2 3 -3 4 -4
```

```python
# Câu 13: Hàm kiểm tra số hoàn thiện, số thịnh vượng
def tinh_tong_uoc_so(n):
    """
    Tính tổng các ước số của số nguyên dương n, không bao gồm n.
    Ví dụ: tinh_tong_uoc_so(6) = 1 + 2 + 3 = 6
    """
    if n <= 0:
        return 0
    
    tong = 0
    for i in range(1, n):
        if n % i == 0:
            tong += i
    return tong

def la_so_hoan_thien(n):
    """
    Kiểm tra xem số nguyên dương n có phải là số hoàn thiện hay không.
    """
    if n <= 0:
        return False
    
    tong_uoc = tinh_tong_uoc_so(n)
    return tong_uoc == n

# Các hàm đã viết ở trên
def tinh_tong_uoc_so(n):
    if n <= 0:
        return 0
    
    tong = 0
    for i in range(1, n):
        if n % i == 0:
            tong += i
    return tong

def la_so_hoan_thien(n):
    if n <= 0:
        return False
    return tinh_tong_uoc_so(n) == n

def la_so_thinh_vuong(n):
    if n <= 0:
        return False
    return tinh_tong_uoc_so(n) > n

# Phần chính của chương trình
try:
    so_nhap = int(input("Nhập một số nguyên dương: "))
    
    if la_so_hoan_thien(so_nhap):
        print(f"Số {so_nhap} là số hoàn thiện.")
    else:
        print(f"Số {so_nhap} không phải là số hoàn thiện.")
        
    if la_so_thinh_vuong(so_nhap):
        print(f"Số {so_nhap} là số thịnh vượng.")
    else:
        print(f"Số {so_nhap} không phải là số thịnh vượng.")

except ValueError:
    print("Đầu vào không hợp lệ. Vui lòng nhập một số nguyên.")
```
