```python
# Câu 1: Xử lý List
from random import randrange

print("Chương trình xử lý List")
n=int(input("Nhập số phần tử"))
lst=[0]*n
for i in range(n):
 lst[i]=randrange(-100,100)
print("List ngẫu nhiên là:")
print(lst)
print("Mời bạn thêm số mới:")
value=int(input())
lst.append(value)
print(lst)
print("bạn muốn đếm số nào:")
k=int(input())
dem=lst.count(k)
print(k,"xuất hiện ",dem,"trong list")
def CheckPrime(n):
 d=0
 for i in range(1,n+1):
  if n%i ==0:
   d+=1
 return d==2
demnt=0
tongnt=0
for x in lst:
 if CheckPrime(x):
  demnt+=1
 tongnt+=x
print("Có ",demnt,"số nguyên tố trong list")
print("Tổng=",tongnt)
lst.sort()
print("List sau khi sort:")
print(lst)
lst.clear()
print("List sau khi xóa:")
print(lst)
```

```python
# Câu 3: Xử lý List Đa chiều
from random import randrange
def TaoMaTran(m,n):
 D=[]
 for i in range(m):
  row=[]
 for j in range(n):
  row.append(randrange(100))
 D.append(row)
 return D
def XuatMaTran(D):
 for row in D:
  for element in row:
   print(element,end='\t')
 print()
def LayDong(r):
 R=D[r]
 return R
def XuatList1Chieu(R):
 for element in R:
  print(element,end='\t')
def LayCot(c):
 C=[]
 for i in range(len(D)):
  C.append(D[i][c])
 return C
def MAX(D):
 max=D[0][0]
 for i in range(len(D)):
  for j in range(len(D[i])):
   if(max<D[i][j]):
    max=D[i][j]
 return max
print("Nhập số dòng:")
m=int(input())
print("Nhập số cột:")
n=int(input())
D=TaoMaTran(m,n)
XuatMaTran(D)
print("Mời bạn nhập dòng muốn xuất:")
r=int(input())
XuatList1Chieu(LayDong(r))
print("Mời bạn nhập cột muốn xuất:")
c=int(input())
XuatList1Chieu(LayCot(c))
max=MAX(D)
print("Số lớn nhất trong ma trận=",max)
```

```python
# Câu 4: Xác định kết quả khi thực thi list
lst = [3, 0, 1, 5, 2]
x = 2

print("Danh sách:", lst)
print("x =", x)
print("---------------------------")

print("1. lst[0] =", lst[0])               # phần tử ở vị trí 0
print("2. lst[3] =", lst[3])               # phần tử ở vị trí 3
print("3. lst[x] =", lst[x])               # x = 2 → phần tử ở vị trí 2
print("4. lst[-x] =", lst[-x])             # -x = -2 → phần tử thứ 2 từ cuối
print("5. lst[x + 1] =", lst[x + 1])       # x + 1 = 3 → phần tử ở vị trí 3
print("6. lst[x] + 1 =", lst[x] + 1)       # lst[x] = 1 → 1 + 1 = 2
print("7. lst[lst[x]] =", lst[lst[x]])     # lst[x] = 1 → lst[1] = 0
print("8. lst[lst[lst[x]]] =", lst[lst[lst[x]]])  # lst[x]=1 → lst[1]=0 → lst[0]=3
```

```python
# Câu 5: Xác định kết quả khi thực thi list
lst = [20, 1, -34, 40, -8, 60, 1, 3]

print("Danh sách ban đầu:", lst)
print("------------------------------------")

print("(a) lst =", lst)
print("(b) lst[0:3] =", lst[0:3])           # Lấy phần tử từ index 0 → 2
print("(c) lst[4:8] =", lst[4:8])           # Lấy phần tử từ index 4 → 7
print("(d) lst[4:33] =", lst[4:33])         # Vượt quá độ dài vẫn an toàn → tới hết list
print("(e) lst[-5:-3] =", lst[-5:-3])       # Đếm từ cuối list: vị trí -5 đến -4
print("(f) lst[-22:3] =", lst[-22:3])       # Chỉ lấy từ đầu tới index 2 vì -22 nhỏ hơn đầu list
print("(g) lst[4:] =", lst[4:])             # Từ index 4 đến hết
print("(h) lst[:] =", lst[:])               # Sao chép toàn bộ list
print("(i) lst[:4] =", lst[:4])             # Từ đầu đến index 3
print("(j) lst[1:5] =", lst[1:5])           # Từ index 1 đến 4
print("(k) -34 in lst =", -34 in lst)       # Kiểm tra -34 có trong list không
print("(l) -34 not in lst =", -34 not in lst) # Kiểm tra -34 không có trong list
print("(m) len(lst) =", len(lst))           # Độ dài list
```

```python
# Câu 6: Nhập vào 1 list có N số ngẫu nhiên KHÔNG TRÙNG NHAU
from random import sample

print("=== CHƯƠNG TRÌNH TẠO LIST NGẪU NHIÊN KHÔNG TRÙNG NHAU ===")
n = int(input("Nhập số phần tử của list: "))

# Dùng sample để lấy ngẫu nhiên n số KHÁC NHAU trong khoảng [-100, 100]
lst = sample(range(-100, 101), n)

print("List ngẫu nhiên KHÔNG TRÙNG LẶP là:")
print(lst)
```

```python
# Câu 7: Viết chương trình nhập vào một dãy các số theo thứ tự tăng, nếu nhập sai
# quy cách thì yêu cầu nhập lại. In dãy số sau khi đã nhập xong 
print("=== CHƯƠNG TRÌNH NHẬP DÃY SỐ TĂNG DẦN ===")

n = int(input("Nhập số phần tử của dãy: "))

while True:
    lst = []
    print("Nhập", n, "số theo thứ tự tăng dần:")
    for i in range(n):
        x = int(input(f"Nhập số thứ {i+1}: "))
        lst.append(x)

    # Kiểm tra xem dãy có tăng dần không
    tang_dan = all(lst[i] < lst[i+1] for i in range(n-1))

    if tang_dan:
        break
    else:
        print("Dãy không hợp lệ! Hãy nhập lại (các số phải tăng dần).\n")

print("Đãy số hợp lệ là:", lst)
```

```python
# Câu 8: Viết chương trình nhập vào một dãy n số thực M[0], M[1],..., M[n-1], sắp xếp
# dãy số theo thứ tự giảm dần. Xuất ra dãy số sau khi sắp xếp.
print("=== CHƯƠNG TRÌNH SẮP XẾP DÃY SỐ GIẢM DẦN ===")

# Nhập số lượng phần tử
n = int(input("Nhập số phần tử của dãy: "))

# Nhập các số thực
M = []
for i in range(n):
    x = float(input(f"Nhập M[{i}]: "))
    M.append(x)

print("\nDãy số ban đầu là:")
print(M)

# Sắp xếp giảm dần
M.sort(reverse=True)

print("\nDãy số sau khi sắp xếp giảm dần là:")
print(M)
```

```python
# Câu 9: Xử lý mảng
print("=== CHƯƠNG TRÌNH XỬ LÝ MẢNG SỐ TỰ NHIÊN ===")

# Có thể cho người dùng nhập hoặc dùng sẵn ví dụ:
# M = list(map(int, input("Nhập các số cách nhau bởi dấu cách: ").split()))
M = [3, 6, 7, 8, 11, 17, 2, 90, 2, 5, 4, 5, 8]

print("Mảng ban đầu:", M)

# --- Hàm kiểm tra số nguyên tố ---
def la_nguyen_to(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

# --- Phân loại ---
le = [x for x in M if x % 2 != 0]
chan = [x for x in M if x % 2 == 0]
ngto = [x for x in M if la_nguyen_to(x)]
kngto = [x for x in M if not la_nguyen_to(x)]

# --- Xuất kết quả ---
print("\n1️ Các số lẻ:", le)
print("Có tổng cộng:", len(le), "số lẻ")

print("\n2️ Các số chẵn:", chan)
print("Có tổng cộng:", len(chan), "số chẵn")

print("\n3️ Các số nguyên tố:", ngto)
print("Có tổng cộng:", len(ngto), "số nguyên tố")

print("\n4️ Các số không phải nguyên tố:", kngto)
print("Có tổng cộng:", len(kngto), "số không nguyên tố")
```

```python
# Câu 10: Xử lý Ma Trận
print("=== CHƯƠNG TRÌNH XỬ LÝ MA TRẬN ===")

# --- Nhập ma trận ---
def nhap_matrix(ten):
    rows = int(input(f"Nhập số dòng của ma trận {ten}: "))
    cols = int(input(f"Nhập số cột của ma trận {ten}: "))
    matrix = []
    print(f"Nhập các phần tử cho ma trận {ten}:")
    for i in range(rows):
        row = list(map(float, input(f"Hàng {i+1}: ").split()))
        while len(row) != cols:
            print("Số phần tử không khớp, nhập lại hàng này!")
            row = list(map(float, input(f"Hàng {i+1}: ").split()))
        matrix.append(row)
    return matrix

# --- Hàm in ma trận ---
def in_matrix(M, name="Matrix"):
    print(f"\n{name}:")
    for row in M:
        print(row)

# --- Hàm cộng 2 ma trận ---
def cong_matrix(A, B):
    rows = len(A)
    cols = len(A[0])
    C = [[A[i][j] + B[i][j] for j in range(cols)] for i in range(rows)]
    return C

# --- Hàm hoán vị (chuyển vị) ma trận ---
def chuyen_vi(M):
    rows = len(M)
    cols = len(M[0])
    T = [[M[j][i] for j in range(rows)] for i in range(cols)]
    return T

# --- Chương trình chính ---
A = nhap_matrix("A")
B = nhap_matrix("B")

# Kiểm tra kích thước trước khi cộng
if len(A) == len(B) and len(A[0]) == len(B[0]):
    C = cong_matrix(A, B)
    in_matrix(A, "Ma trận A")
    in_matrix(B, "Ma trận B")
    in_matrix(C, "Tổng A + B")
else:
    print("Hai ma trận không cùng kích thước, không thể cộng!")

# --- Tìm ma trận chuyển vị ---
TA = chuyen_vi(A)
TB = chuyen_vi(B)

in_matrix(TA, "Chuyển vị của A")
in_matrix(TB, "Chuyển vị của B")
```
